```python
plt.figure(figsize=(6,4))
sns.countplot(x='Expert Diagnose', data=M, palette='coolwarm')
plt.title('Distribution of Diagnoses')
plt.show()
```

Each bar's height corresponds to the number of people diagnosed with that condition.Bipolar Type-2 and Depression have the highest counts (around 31 each).Bipolar Type-1 has a slightly lower count (around 28).Normal individuals are close behind with around 30.

```python
binary_cols = ['Mood Swing', 'Suicidal thoughts', 'Anorxia', 'Authority Respect', 'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On', 'Nervous Break-down', 'Admit Mistakes', 'Overthinking']

plt.figure(figsize=(12,6))
for i, col in enumerate(binary_cols[:6], 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=col, hue='Expert Diagnose', data=M, palette='coolwarm')
    plt.title(col)
plt.tight_layout()
plt.show()
```

Mood Swing

People with Bipolar Type-1 and Type-2 report mood swings more often, while people with Depression and those in the Normal group usually do not.

Suicidal Thoughts

People in the Bipolar Type-2 and Depression groups report suicidal thoughts more often, while most people in the Normal group do not.

Anorexia

Most people in all groups, especially in the Normal group, do not report anorexia, although a few people in each group do.

Authority Respect

YES responses are highest in the Normal group.NO responses are more common in Bipolar and Depression groups.

Try-Explanation

The responses are fairly balanced, but slightly more people in every group tend to try to explain themselves.

Aggressive Response

NO dominates in the Normal and Depression groups.Bipolar Type-1 and Type-2 show higher YES responses.

```python
plt.figure(figsize=(5,5))
M["Expert Diagnose"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("coolwarm", 5))
plt.title("Percentage of Each Diagnosis")
plt.ylabel("")
plt.show()
```

percentage of each diagnosis patients.

```python
mood_features = ['Sadness', 'Euphoric', 'Exhausted', 'Sleep dissorder', 'Mood Swing']
plt.figure(figsize=(10,6))
for feature in mood_features:
    sns.countplot(x=feature, data=M)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(M.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Mental Health Variables")
plt.show()
```

Sexual Activity & Concentration-Very weak negative correlation(-0.082) these two are almost unrelated.Sexual Activity & Optimism-Moderate positive correlation (0.56) higher sexual activity tends to be associated with higher optimism.Concentration & Optimism-Very weak positive correlation (0.12)nearly no relationship.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Overthinking", y="Concentration", hue="Expert Diagnose", data=M, s=100)
plt.title("Overthinking vs Concentration Levels")
plt.show()
```

```python
binary_features = ['Suicidal thoughts', 'Anorxia', 'Authority Respect',
                   'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On',
                   'Nervous Break-down', 'Admit Mistakes', 'Overthinking']

for feature in binary_features:
    plt.figure(figsize=(6,4))
    sns.countplot(x=feature, hue='Expert Diagnose', data=M, palette='Set3')
    plt.title(f"{feature} by Diagnosis Type")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title='Diagnosis')
    plt.show()
```
